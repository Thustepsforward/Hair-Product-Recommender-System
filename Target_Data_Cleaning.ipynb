## **Hair Styling Products**

```python
Target_hsp = pd.read_csv('/content/drive/MyDrive/Capstone Project - Summer 2024/combined_Target_hair_styling_products.csv')
# Remove unwanted characters and convert to int
Target_hsp['Number of Ratings'] = Target_hsp['Number of Ratings'].str.replace(' reviews', '').astype(int)
# Extract ratings as a float. '\d+' captures one or more digits; '\.?' captures an optional decimal point; '\d*' captures any number of digits following the decimal point, could be none
Target_hsp['Rating'] = Target_hsp['Rating'].str.extract('(\d+\.?\d*)').astype(float)
# Drop duplicates
Target_hsp = Target_hsp.drop_duplicates().reset_index(drop=True)
Target_hsp
```

### **Filtering Decisions**

**Filter the Dataset Based on Criteria**

```python
Target_hsp = Target_hsp[(Target_hsp['Number of Ratings'] > 100) & (Target_hsp['Rating'] >= 4.5)]
Target_hsp.reset_index(drop=True, inplace = True)
Target_hsp
```

**Remove any Kits and Duos from the dataset.**

```python
Target_hsp = Target_hsp[~Target_hsp['Product Name'].str.contains(' Kit| Duo|Shampoo and Conditioner', case = False, na=False)]
Target_hsp.reset_index(drop=True, inplace = True)
Target_hsp
```

**Remove any Kids' or Men products.**

```python
Target_hsp = Target_hsp[~Target_hsp['Product Name'].str.contains(' kids|baby| men', case = False, na=False)]
Target_hsp = Target_hsp[~Target_hsp['Brand'].str.contains('Fairy Tales', case = False, na=False)]
Target_hsp.reset_index(drop=True, inplace = True)
Target_hsp
```

**Remove any skin or body, pre wash, and heat protectant products.**

```python
Target_hsp = Target_hsp[~Target_hsp['Product Name'].str.contains('body wash| skin| heat protect| hair spray| hairspray|pre wash', case = False, na=False)]
Target_hsp.reset_index(drop=True, inplace = True)
Target_hsp
```
